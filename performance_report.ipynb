# Relatório de Performance

## Complexidade Teórica: Análise do custo O(n) para inserção/busca em listas encadeadas

### Introdução

As estruturas de dados lineares implementadas neste projeto (Fila, Lista Duplamente Encadeada e Pilha) apresentam características distintas de complexidade. Enquanto Fila e Pilha garantem operações O(1) nas extremidades, a **Lista Duplamente Encadeada Ordenada** utilizada no Livro de Ofertas apresenta complexidade O(n) para inserção e busca.

### 1. Busca em Lista Encadeada - Complexidade O(n)

Em uma lista duplamente encadeada, **não existe acesso direto** aos elementos como em *arrays*. Para localizar um elemento ou encontrar a posição de inserção ordenada, é necessário **percorrer sequencialmente** desde o *head* (ou *tail*) até encontrar o nó desejado.

#### Operações por caso:

- **Início da lista**: 1 comparação
- **Meio da lista**: n/2 comparações  
- **Fim da lista**: n comparações
- **Elemento não encontrado**: n comparações (necessário verificar toda a lista)

**Complexidade média**: O(n/2) = **O(n)**  
**Complexidade pior caso**: O(n)

### 2. Inserção Ordenada em Lista Encadeada - Complexidade O(n)

A operação de inserção ordenada envolve duas etapas:

#### Etapa 1: Busca da posição correta
Para manter a lista ordenada, é necessário:
1. Percorrer a lista a partir do head
2. Comparar o preço da ordem com cada nó existente
3. Parar quando encontrar a posição correta

**Custo**: O(n) comparações em média

#### Etapa 2: Ajuste de ponteiros
Após encontrar a posição correta:
- Criar novo nó: O(1)
- Ajustar referências next/prev: O(1)
- Total desta etapa: O(1)

**Complexidade total de inserção**: O(n) + O(1) = **O(n)**

### 3. Operações nas extremidades em Fila (*Queue*) e Pilha (*Stack*) - Complexidade O(1)

Diferentemente da lista ordenada, essas estruturas garantem operações rápidas nas extremidades:
- **_Enqueue_**: Adiciona ao *tail* → O(1)
- **_Dequeue_**: Remove do *head* → O(1)
- **_Push_**: Adiciona ao topo → O(1)
- **_Pop_**: Remove do topo → O(1)

### 4. Comparação de Complexidades

| Operação | Lista Ordenada | Fila | Pilha |
|----------|----------------|------|-------|
| Busca | O(n) | O(n) | O(n) |
| Inserção (ordenada) | O(n) | O(1) | O(1) |
| Remoção (específica) | O(n) | - | - |
| Acesso ao melhor | O(1) | O(1) | O(1) |
