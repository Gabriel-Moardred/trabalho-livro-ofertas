## Complexidade Teórica: Análise do custo O(n) para inserção/busca em listas encadeadas

### Introdução

As estruturas de dados lineares implementadas neste projeto (Fila, Lista Duplamente Encadeada e Pilha) apresentam características distintas de complexidade. Enquanto Fila e Pilha garantem operações O(1) nas extremidades, a **Lista Duplamente Encadeada Ordenada** utilizada no Livro de Ofertas apresenta complexidade O(n) para inserção e busca.

### 1. Busca em Lista Encadeada - Complexidade O(n)

Em uma lista duplamente encadeada, **não existe acesso direto** aos elementos como em *arrays*. Para localizar um elemento ou encontrar a posição de inserção ordenada, é necessário **percorrer sequencialmente** desde o *head* (ou *tail*) até encontrar o nó desejado.

#### Operações por caso:

- **Início da lista**: 1 comparação
- **Meio da lista**: n/2 comparações  
- **Fim da lista**: n comparações
- **Elemento não encontrado**: n comparações (necessário verificar toda a lista)

**Complexidade média**: O(n/2) = **O(n)**  
**Complexidade pior caso**: O(n)

### 2. Inserção Ordenada em Lista Encadeada - Complexidade O(n)

A operação de inserção ordenada envolve duas etapas:

#### Etapa 1: Busca da posição correta
Para manter a lista ordenada, é necessário:
1. Percorrer a lista a partir do head
2. Comparar o preço da ordem com cada nó existente
3. Parar quando encontrar a posição correta

**Custo**: O(n) comparações em média

#### Etapa 2: Ajuste de ponteiros
Após encontrar a posição correta:
- Criar novo nó: O(1)
- Ajustar referências next/prev: O(1)
- Total desta etapa: O(1)

**Complexidade total de inserção**: O(n) + O(1) = **O(n)**

### 3. Operações nas extremidades em Fila (*Queue*) e Pilha (*Stack*) - Complexidade O(1)

Diferentemente da lista ordenada, essas estruturas garantem operações rápidas nas extremidades:
- **_Enqueue_**: Adiciona ao *tail* → O(1)
- **_Dequeue_**: Remove do *head* → O(1)
- **_Push_**: Adiciona ao topo → O(1)
- **_Pop_**: Remove do topo → O(1)

### 4. Comparação de Complexidades

| Operação | Lista Ordenada | Fila | Pilha |
|----------|----------------|------|-------|
| Busca | O(n) | O(n) | O(n) |
| Inserção (ordenada) | O(n) | O(1) | O(1) |
| Remoção (específica) | O(n) | - | - |
| Acesso ao melhor | O(1) | O(1) | O(1) |

### 5. Impacto no Livro de Ofertas

No simulador de livro de ofertas:

**Livro de Compras** (ordem decrescente de preço):
- Inserção de nova ordem: percorre até encontrar posição → O(n)

**Livro de Vendas** (ordem crescente de preço):
- Inserção de nova ordem: percorre até encontrar posição → O(n)

**Motor de Match**:
- Verifica melhor compra (head): O(1)
- Verifica melhor venda (head): O(1)
- Match em si: O(1) ou O(n) dependendo da quantidade removida

### 8. Fluxo Completo do Sistema e Custos

No motor de negociação:

```
Ordem entra → Fila FIFO [O(1) enqueue]
             ↓
        Dequeue [O(1)]
             ↓
    Inserir em Lista Ordenada [O(n) ← GARGALO]
             ↓
    Motor de Match (comparação heads) [O(1)]
             ↓
    Remover nós casados [O(n) ou O(1) dependendo]
             ↓
    Empilhar undo [O(1) push]
```

**O gargalo da performance está na inserção ordenada com complexidade O(n).**

````python
import time
import random
import matplotlib.pyplot as plt

# Importando as estruturas do seu próprio arquivo structs.py
from structs import Order, DoublyLinkedList

def medir_tempo_insercao(volumes):
    tempos = []
    
    for n in volumes:
        livro_compras = DoublyLinkedList()
        inicio = time.time() # Inicia o cronômetro
        
        for i in range(n):
            # Gerando um preço aleatório para forçar a busca na lista (O(n))
            preco = random.uniform(10.0, 100.0)
            qtd = random.randint(10, 500)
            nova_ordem = Order(i, 'C', preco, qtd)
            
            # O gargalo de performance identificado teoricamente
            livro_compras.insert_ordered(nova_ordem, is_buy_list=True)
            
        fim = time.time() # Para o cronômetro
        tempo_total = fim - inicio
        tempos.append(tempo_total)
        print(f"Volume: {n} ordens | Tempo: {tempo_total:.4f} segundos")
        
    return tempos

# Definindo cenários de teste progressivos
cenarios_volume = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]

print("Iniciando bateria de testes empíricos...")
tempos_execucao = medir_tempo_insercao(cenarios_volume)

# Gerando o gráfico exigido pelo trabalho
plt.figure(figsize=(10, 6))
plt.plot(cenarios_volume, tempos_execucao, marker='o', linestyle='-', color='purple', linewidth=2)
plt.fill_between(cenarios_volume, tempos_execucao, color='purple', alpha=0.1)
plt.title("Análise Empírica: Tempo de Inserção x Volume de Ordens", fontsize=14, fontweight='bold')
plt.xlabel("Volume de Ordens Inseridas (N)", fontsize=12)
plt.ylabel("Tempo Total de Execução (Segundos)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Exibindo o gráfico
plt.show()
````
